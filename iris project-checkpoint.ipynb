```python
def get_sepal_size(sepal_area):
  if sepal_area > 2:
    return "large"
  else:
    return "small"

df["sepal_size"] = df["sepal_area"].apply(get_sepal_size)
df.head()
```

```python
df["sepal_size"].value_counts()
```

```python
df.groupby(["species","sepal_size"])["petal_width"].mean()
```
